```python
import torch
from transformers import AutoTokenizer, AutoModelForQuestionAnswering

# Load the model and tokenizer
model_name = "bert-large-uncased-whole-word-masking-finetuned-squad"  # You can replace this with other models like "bert-large-uncased-whole-word-masking-finetuned-squad"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForQuestionAnswering.from_pretrained(model_name)

# Define context and question
context = "Claude is a large language model developed by Anthropic, designed to be safe and interpretable."
question = "Who developed Claude?"

# Tokenize input
inputs = tokenizer(question, context, return_tensors='pt')
input_ids = inputs["input_ids"].tolist()[0]

# Get model outputs
outputs = model(**inputs)
answer_start_scores = outputs.start_logits
answer_end_scores = outputs.end_logits

# Get the most likely beginning and end of answer
answer_start = torch.argmax(answer_start_scores)  # Get the most likely beginning of answer
answer_end = torch.argmax(answer_end_scores) + 1  # Get the most likely end of answer

# Convert tokens to answer
answer = tokenizer.convert_tokens_to_string(tokenizer.convert_ids_to_tokens(input_ids[answer_start:answer_end]))
print(f"Question: {question}\nAnswer: {answer}")
```
