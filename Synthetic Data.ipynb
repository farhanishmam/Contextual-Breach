```python
import random
import nltk
from nltk.corpus import words
import string

nltk.download('punkt')
nltk.download('words')
```

```python
class TextPerturbator:
    def __init__(self):
        self.random_words_list = words.words()

        self.qwerty_neighbors = {
            'a': 'qwsz',
            'b': 'vghn',
            'c': 'xdfv',
            'd': 'serfcx',
            'e': 'wsdr',
            'f': 'rtgvcx',
            'g': 'tyhbv',
            'h': 'yujnbg',
            'i': 'ujko',
            'j': 'uikmnh',
            'k': 'iolmj',
            'l': 'opk',
            'm': 'njk',
            'n': 'bhjm',
            'o': 'iklp',
            'p': 'ol',
            'q': 'wa',
            'r': 'edft',
            's': 'awedxz',
            't': 'rfgy',
            'u': 'yhji',
            'v': 'cfgb',
            'w': 'qase',
            'x': 'zsdc',
            'y': 'tghu',
            'z': 'asx',
            'A': 'QWSZ',
            'B': 'VGHN',
            'C': 'XDFV',
            'D': 'SERFCX',
            'E': 'WSDR',
            'F': 'RTGVCX',
            'G': 'TYHBV',
            'H': 'YUJNBG',
            'I': 'UJKO',
            'J': 'UIKMNH',
            'K': 'IOLMJ',
            'L': 'OPK',
            'M': 'NJK',
            'N': 'BHJM',
            'O': 'IKLP',
            'P': 'OL',
            'Q': 'WA',
            'R': 'EDFT',
            'S': 'AWEDXZ',
            'T': 'RFGY',
            'U': 'YHJI',
            'V': 'CFGB',
            'W': 'QASE',
            'X': 'ZSDC',
            'Y': 'TGHU',
            'Z': 'ASX'
        }

    def add_random_perturbations(self, paragraph, level):
        if level < 1 or level > 5:
            raise ValueError("Level must be between 1 and 5")

        sentences = nltk.sent_tokenize(paragraph)
        severity_ratio = level / 5

        new_sentences = []

        for sentence in sentences:
            words_in_sentence = nltk.word_tokenize(sentence)
            num_perturbations = int(len(words_in_sentence) * severity_ratio)

            for _ in range(num_perturbations):
                word_index = random.randint(0, len(words_in_sentence) - 1)
                word = words_in_sentence[word_index]

                if len(word) > 1:  # To ensure there's space to add perturbations
                    if random.choice([True, False]):
                        # Add a space at a random position
                        position = random.randint(0, len(word))
                        word = word[:position] + ' ' + word[position:]
                    else:
                        # Add a random character at a random position
                        random_char = random.choice(string.ascii_letters)
                        position = random.randint(0, len(word))
                        word = word[:position] + random_char + word[position:]

                    # Update the word in the list
                    words_in_sentence[word_index] = word

            new_sentences.append(' '.join(words_in_sentence))

        return ' '.join(new_sentences)


    def add_typing_mistakes(self, paragraph, level):
        if level < 1 or level > 5:
            raise ValueError("Level must be between 1 and 5")

        sentences = nltk.sent_tokenize(paragraph)
        severity_ratio = level / 5

        new_sentences = []

        for sentence in sentences:
            words_in_sentence = nltk.word_tokenize(sentence)
            num_mistakes = int(len(words_in_sentence) * severity_ratio)

            for _ in range(num_mistakes):
                word_index = random.randint(0, len(words_in_sentence) - 1)
                word = words_in_sentence[word_index]

                if len(word) > 1:  # Ensure the word has more than one character
                    char_index = random.randint(0, len(word) - 1)

                    if random.choice([True, False]):
                        # Replace with a neighboring character on the keyboard
                        char = word[char_index]
                        if char in self.qwerty_neighbors:
                            neighbor_char = random.choice(self.qwerty_neighbors[char])
                            word = word[:char_index] + neighbor_char + word[char_index + 1:]
                    else:
                        # Swap with an adjacent character
                        if char_index < len(word) - 1:
                            word = (
                                word[:char_index] +
                                word[char_index + 1] +
                                word[char_index] +
                                word[char_index + 2:]
                            )

                    # Update the word in the list
                    words_in_sentence[word_index] = word

            new_sentences.append(' '.join(words_in_sentence))

        return ' '.join(new_sentences)


    def delete_characters(self, paragraph, level):
            if level < 1 or level > 5:
                raise ValueError("Level must be between 1 and 5")

            sentences = nltk.sent_tokenize(paragraph)
            severity_ratio = level / 5

            new_sentences = []

            for sentence in sentences:
                words_in_sentence = nltk.word_tokenize(sentence)
                num_deletions = int(len(words_in_sentence) * severity_ratio)

                for _ in range(num_deletions):
                    word_index = random.randint(0, len(words_in_sentence) - 1)
                    word = words_in_sentence[word_index]

                    if len(word) > 1:  # Ensure the word has more than one character
                        char_index = random.randint(0, len(word) - 1)
                        word = word[:char_index] + word[char_index + 1:]

                        # Update the word in the list
                        words_in_sentence[word_index] = word

                new_sentences.append(' '.join(words_in_sentence))

            return ' '.join(new_sentences)
```

```python
import random
import nltk
from nltk.corpus import wordnet, words
import string

nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('words')
```

```python
class WordPerturbator:
    def __init__(self):
        self.random_words_list = words.words()

    def get_synonyms(self, word, pos):
        synonyms = set()
        for syn in wordnet.synsets(word, pos=pos):
            for lemma in syn.lemmas():
                if lemma.name().lower() != word.lower():
                    synonyms.add(lemma.name().replace('_', ' '))
        return list(synonyms)

    def pos_tag_to_wordnet_tag(self, tag):
        if tag.startswith('J'):
            return wordnet.ADJ
        elif tag.startswith('V'):
            return wordnet.VERB
        elif tag.startswith('N'):
            return wordnet.NOUN
        elif tag.startswith('R'):
            return wordnet.ADV
        else:
            return None

    def is_proper_noun(self, tag):
        return tag == 'NNP' or tag == 'NNPS'

    def perturb_with_synonyms(self, paragraph, level):
        if level < 1 or level > 5:
            raise ValueError("Level must be between 1 and 5")

        severity_ratio = level / 5

        sentences = nltk.sent_tokenize(paragraph)
        new_sentences = []

        for sentence in sentences:
            words_in_sentence = nltk.word_tokenize(sentence)
            tagged_words = nltk.pos_tag(words_in_sentence)

            num_perturbations = int(len(words_in_sentence) * severity_ratio)
            perturbation_indices = random.sample(range(len(words_in_sentence)), num_perturbations)

            for i in perturbation_indices:
                word, tag = tagged_words[i]
                if self.is_proper_noun(tag):
                    continue

                wn_tag = self.pos_tag_to_wordnet_tag(tag)
                if wn_tag:
                    synonyms = self.get_synonyms(word, wn_tag)
                    if synonyms:
                        words_in_sentence[i] = random.choice(synonyms)

            new_sentences.append(' '.join(words_in_sentence))

        return ' '.join(new_sentences)

    def random_insertion(self, paragraph, level):
        if level < 1 or level > 5:
            raise ValueError("Level must be between 1 and 5")

        severity_ratio = level / 5

        sentences = nltk.sent_tokenize(paragraph)
        new_sentences = []

        for sentence in sentences:
            words_in_sentence = nltk.word_tokenize(sentence)
            num_insertions = int(len(words_in_sentence) * severity_ratio)

            for _ in range(num_insertions):
                random_word = random.choice(self.random_words_list)
                insert_position = random.randint(0, len(words_in_sentence))
                words_in_sentence.insert(insert_position, random_word)

            new_sentences.append(' '.join(words_in_sentence))

        return ' '.join(new_sentences)

    def reorder_words(self, paragraph, level):
        if level < 1 or level > 5:
            raise ValueError("Level must be between 1 and 5")

        severity_ratio = level / 5

        sentences = nltk.sent_tokenize(paragraph)
        new_sentences = []

        for sentence in sentences:
            words_in_sentence = nltk.word_tokenize(sentence)
            num_reorders = int(len(words_in_sentence) * severity_ratio)

            for _ in range(num_reorders):
                if len(words_in_sentence) > 1:
                    idx1, idx2 = random.sample(range(len(words_in_sentence)), 2)
                    words_in_sentence[idx1], words_in_sentence[idx2] = words_in_sentence[idx2], words_in_sentence[idx1]

            new_sentences.append(' '.join(words_in_sentence))

        return ' '.join(new_sentences)
```

```python
paragraph = "John went to the market to buy fresh apples and bananas. He loves fruits."
level = 3

word_perturbator = WordPerturbator()
perturbed_paragraph = word_perturbator.perturb_with_synonyms(paragraph, level)
print("Perturbed Paragraph with synonyms:")
print(perturbed_paragraph)
```

```python
import random
import nltk
from nltk.corpus import wordnet, words

nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('words')

class WordPerturbator:
    def __init__(self):
        self.random_words_list = words.words()

    def pos_tag_to_wordnet_tag(self, tag):
        if tag.startswith('J'):
            return wordnet.ADJ
        elif tag.startswith('V'):
            return wordnet.VERB
        elif tag.startswith('N'):
            return wordnet.NOUN
        elif tag.startswith('R'):
            return wordnet.ADV
        else:
            return None

    def is_proper_noun(self, tag):
        return tag == 'NNP' or tag == 'NNPS'

    def introduce_grammatical_mistakes(self, paragraph, level):
        if level < 1 or level > 5:
            raise ValueError("Level must be between 1 and 5")

        severity_ratio = level / 5
        sentences = nltk.sent_tokenize(paragraph)
        new_sentences = []

        for sentence in sentences:
            words_in_sentence = nltk.word_tokenize(sentence)
            tagged_words = nltk.pos_tag(words_in_sentence)
            num_mistakes = int(len(words_in_sentence) * severity_ratio)

            for _ in range(num_mistakes):
                mistake_type = random.choice(["verb_tense", "subject_verb_agreement", "preposition", "article"])

                if mistake_type == "verb_tense":
                    for i, (word, tag) in enumerate(tagged_words):
                        if tag.startswith('V'):
                            new_word = random.choice(self.random_words_list)
                            words_in_sentence[i] = new_word
                            break

                elif mistake_type == "subject_verb_agreement":
                    for i, (word, tag) in enumerate(tagged_words):
                        if tag in ['VBZ', 'VBP']:
                            if tag == 'VBZ':
                                words_in_sentence[i] = random.choice(["are", "were"])
                            elif tag == 'VBP':
                                words_in_sentence[i] = random.choice(["is", "was"])
                            break

                elif mistake_type == "preposition":
                    prepositions = ["in", "on", "at", "by", "with", "about", "against", "between", "into", "through"]
                    for i, (word, tag) in enumerate(tagged_words):
                        if tag == 'IN' and word in prepositions:
                            words_in_sentence[i] = random.choice(prepositions)
                            break

                elif mistake_type == "article":
                    articles = ["a", "an", "the"]
                    for i, (word, tag) in enumerate(tagged_words):
                        if tag == 'DT' and word in articles:
                            words_in_sentence[i] = random.choice(articles)
                            break

            new_sentences.append(' '.join(words_in_sentence))

        return ' '.join(new_sentences)
```

```python
import random
import nltk
from nltk.corpus import wordnet

nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')
nltk.download('omw-1.4')

class SentencePerturbator:
    def __init__(self):
        pass

    def introduce_grammatical_mistakes(self, paragraph, level):
        if level < 1 or level > 5:
            raise ValueError("Level must be between 1 and 5")

        severity_ratio = level / 5
        sentences = nltk.sent_tokenize(paragraph)
        new_sentences = []

        for sentence in sentences:
            words_in_sentence = nltk.word_tokenize(sentence)
            tagged_words = nltk.pos_tag(words_in_sentence)
            num_mistakes = int(len(words_in_sentence) * severity_ratio)

            for _ in range(num_mistakes):
                mistake_type = random.choice(["verb_tense", "subject_verb_agreement", "preposition", "article", "double_negation", "misplaced_modifier"])

                if mistake_type == "verb_tense":
                    for i, (word, tag) in enumerate(tagged_words):
                        if tag.startswith('V'):
                            # Introduce a verb tense error
                            if tag in ["VBD", "VBN"]:
                                words_in_sentence[i] = random.choice(["is", "am", "are", "was", "were"])
                            else:
                                words_in_sentence[i] = random.choice(["had", "did"])
                            break

                elif mistake_type == "subject_verb_agreement":
                    for i, (word, tag) in enumerate(tagged_words):
                        if tag in ['VBZ', 'VBP']:
                            # Introduce a subject-verb agreement error
                            if tag == 'VBZ':
                                words_in_sentence[i] = random.choice(["are", "were"])
                            elif tag == 'VBP':
                                words_in_sentence[i] = random.choice(["is", "was"])
                            break

                elif mistake_type == "preposition":
                    prepositions = ["in", "on", "at", "by", "with", "about", "against", "between", "into", "through"]
                    for i, (word, tag) in enumerate(tagged_words):
                        if tag == 'IN' and word in prepositions:
                            # Misuse of prepositions
                            words_in_sentence[i] = random.choice(prepositions)
                            break

                elif mistake_type == "article":
                    articles = ["a", "an", "the"]
                    for i, (word, tag) in enumerate(tagged_words):
                        if tag == 'DT' and word in articles:
                            # Incorrect article usage
                            words_in_sentence[i] = random.choice(articles)
                            break

                elif mistake_type == "double_negation":
                    for i, (word, tag) in enumerate(tagged_words):
                        if tag == 'RB' and word in ["not", "never"]:
                            # Introduce a double negation
                            words_in_sentence.insert(i + 1, random.choice(["not", "never"]))
                            break

                elif mistake_type == "misplaced_modifier":
                    if len(words_in_sentence) > 2:
                        idx = random.randint(1, len(words_in_sentence) - 2)
                        words_in_sentence[idx], words_in_sentence[idx + 1] = words_in_sentence[idx + 1], words_in_sentence[idx]

            new_sentences.append(' '.join(words_in_sentence))

        return ' '.join(new_sentences)
```

```python
sentence_perturbator = SentencePerturbator()
for l in range(1,6):
  new_df = df.copy()
  for i in tqdm(range(len(df))):
    for j in range(len(df.iloc[i]['data']['paragraphs'])):
      context = df.iloc[i]['data']['paragraphs'][j]['context']
      new_context = sentence_perturbator.introduce_grammatical_mistakes(context, l) ## This is where the functions are called.
      new_df.iloc[i]['data']['paragraphs'][j]['new_context'] = new_context

  json_name = 'grammatical_mistakes_sentence'+'_'+str(l)+'.json' ## Name the json file, here it says insert_chars
  new_df.to_json(json_name, orient='records')

  from google.colab import files
  files.download(json_name)
```
