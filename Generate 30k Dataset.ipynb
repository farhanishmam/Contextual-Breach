```python
def delete_is_impossible(data):
  for article in data['data']:
      # Iterate over each paragraph in the article
      for paragraph in article['paragraphs']:
          # Filter the QA pairs where 'is_impossible' is False
          paragraph['qas'] = [qa for qa in paragraph['qas'] if not qa['is_impossible']]

  with open('temp.json', 'w', encoding='utf-8') as file:
    json.dump(data, file, ensure_ascii=False, indent=4)
```

```python
def load_squad_dataset(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        squad_data = json.load(file)
    return squad_data

# Save the subset dataset to a new JSON file
def save_squad_dataset(sampled_data, output_file_path):
    with open(output_file_path, 'w', encoding='utf-8') as file:
        json.dump(sampled_data, file, ensure_ascii=False, indent=4)
```

```python
def main():
    # Load the original SQuAD dataset

    noise_list = ["delete_character","grammatical_mistakes_sentence", "grammatical_mistakes_word", "insert_chars", "random_insertion", "reorder_words", "synonym_replacement", "typing_mistake"]

    for noise in noise_list:
      print(noise)
      for level in (range(1, 6)):
        input_file_path = f'/content/drive/MyDrive/Research/Robustness of QA/Reformed Dataset/{noise}/{noise}_{level}.json'

        delete_is_impossible(load_squad_dataset(input_file_path))

        input_file_path = 'temp.json'

        squad_data = load_squad_dataset(input_file_path)

        # Extract the first 100 entries
        first_100_data = squad_data['data'][:150]

        # Create a new dictionary for the sampled data
        sampled_data = {
            'version': squad_data['version'],
            'data': first_100_data
        }
        output_file_path = f'/content/drive/MyDrive/Research/Robustness of QA/Reformed Dataset 30k/{noise}/{noise}_{level}.json'
        save_squad_dataset(sampled_data, output_file_path)
```

```python
if __name__ == "__main__":
    main()
```
